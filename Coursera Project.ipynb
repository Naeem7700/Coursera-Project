```python
import pandas as pd
import numpy as np 
import random 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Generating Data

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
n = 500
data_set = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}

# Create a DataFrame from the data dictionary
df = pd.DataFrame(data_set)

# Print the first few rows of the DataFrame to view the generated data
df.head(10)

```

# Unique Values in Category 

```python
unique_values=df['Category'].nunique()
print(unique_values)
```

```python
counts=df.value_counts('Category', ascending=True)
print(counts)
```

# Missing Values 

```python
missing_values = df.isnull().sum()
print(missing_values)
```

This implies the data has no missing values.

# Duplicate Values

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
```

# Outliers 

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.xticks(rotation=45)
plt.title('Likes Distribution by Category')
plt.show()
```

This Implies no outliers in the data

# Insights Through Graphs 

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Likes', data=df)
plt.xticks(rotation=45)
plt.title('Likes by Category')
plt.show()
```

```python
likes_by_category = df.groupby('Category')['Likes'].sum()
plt.figure(figsize=(6, 6))
plt.pie(likes_by_category, labels=likes_by_category.index, autopct='%1.1f%%')
plt.title('Proportion of Likes by Category')
plt.show()
```
